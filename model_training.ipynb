**Evaluate and predict**

```python
from ultralytics import YOLO
import os, pandas as pd

# --- Paths
BEST = "/content/drive/MyDrive/runs_car_damage/v8n_896/weights/best.pt"
DATA = "/content/dataset/data.yaml"
IMGSZ = 896
DEVICE = 0
PROJECT = "/content/drive/MyDrive/runs_car_damage"  # where to save outputs
RUN_NAME = "eval_test_v8n_896"                      # folder name for this eval

# 1) Load model
model = YOLO(BEST)
print("Loaded:", BEST)

# 2) Evaluate on TEST split and save plots + COCO-style JSON
metrics = model.val(
    data=DATA,
    split="test",
    imgsz=IMGSZ,
    device=DEVICE,
    plots=True,
    save_json=True,
    project=PROJECT,
    name=RUN_NAME
)

print("\n=== TEST Evaluation ===")
print(f"mAP50-95: {metrics.box.map:.3f}")
print(f"mAP50   : {metrics.box.map50:.3f}")
print("Results saved to:", metrics.save_dir)

# 3) Visual predictions on TEST images for qualitative check
#    Save drawn images (with boxes) to PROJECT/predict_test_v8s_832
PRED_SOURCE = "/content/dataset/images/test"
pred = model.predict(
    source=PRED_SOURCE,
    imgsz=IMGSZ,
    device=DEVICE,
    conf=0.25,
    save=True,
    project=PROJECT,
    name="predict_test_v8n_896"
)
print("Predictions saved to:", pred[0].save_dir)
```
